```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Q1_2017 = pd.read_csv("Citibike_data/2017Q1.csv")
Q2_2017 = pd.read_csv("Citibike_data/2017Q2.csv")
Q3_2017 = pd.read_csv("Citibike_data/2017Q3.csv")
Q4_2017 = pd.read_csv("Citibike_data/2017Q4.csv")
```

```python
dataframes = [Q1_2017, Q2_2017, Q3_2017, Q4_2017]
merged_citibike = pd.concat(dataframes)
merged_citibike = merged_citibike.reset_index(drop = True)
merged_citibike.to_csv("Citibike_data/merged_citibike.csv")
merged_citibike
```
